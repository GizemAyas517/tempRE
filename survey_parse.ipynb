```python
import pandas as pd
ex1 = pd.read_excel("User Stories 1-2.xlsx")
ex2 = pd.read_excel("User Stories 2-2.xlsx")
ex3 = pd.read_excel("User Stories 3-2.xlsx")
ex4 = pd.read_excel("User Stories 4-2.xlsx")
ex1.columns = ["text", "pronouns", "noun1", "noun2", "noun3"]
ex2.columns = ["text", "pronouns", "noun1", "noun2", "noun3"]
ex3.columns = ["text", "pronouns", "noun1", "noun2", "noun3"]
ex4.columns = ["text", "pronouns", "noun1", "noun2", "noun3"]
for i in range(49):
  print(ex1.loc[i].text)
  for index,noun in enumerate(ex1.loc[i].pronouns[1:-1].replace("\"","").replace("'","").split(",")):
      print("noun= "+ noun, end="\t")
      print(" %s" %ex1.loc[i]["noun"+ str(index+1)],end="; ")
      print(" %s" %ex2.loc[i]["noun"+ str(index+1)],end="; ")
      print(" %s" %ex3.loc[i]["noun"+ str(index+1)],end="; ")
      print(" %s" %ex4.loc[i]["noun"+ str(index+1)],end="; ")
      print(" ")
     
    
  print("\n--------------------\n")
```
